```python
import numpy as np

# Parameters
num_samples = 6000
sample_rate = 44100  # Hz
frequency = 140      # Hz (A4 note)

# Generate time array
t = np.linspace(0, 6000, num_samples)

# Create sine wave
sine_wave = np.sin(2 * np.pi * frequency * t)

print(sine_wave[2])
print(sine_wave[:4])
```

```python

# Optional: Plot the first 100 samples
import matplotlib.pyplot as plt
plt.figure(figsize=(50, 4))
plt.plot(t[:6000], sine_wave[:6000])
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.grid(True)
plt.show()
```

```python

from datetime import datetime


array_x = []


# indices = np.arange(0, lengthOfRow, 100)
for idx in range(0,6000):
    # print(f"Value at position {idx}: {df[idx]}")
    specific_timestamp = idx
    date_from_timestamp = datetime.fromtimestamp(specific_timestamp)
    array_x.append(date_from_timestamp)


print(array_x[:4])
print(len(array_x))
```

```python
import csv

# Create and write to CSV file
with open('sin-with-date.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    # Write headers
    writer.writerow(['date', 'pressure'])
    # Write data rows
    writer.writerows(zip(array_x, sine_wave))
```
