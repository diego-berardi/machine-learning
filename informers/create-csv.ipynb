```python
import scipy.io as sio
import pandas as pd
import numpy as np
```

```python

mat_data = sio.loadmat('../B101/Channel1.mat')
print(mat_data.keys())
x = mat_data['singleData']
print(x)
```

```python
df = pd.DataFrame(x)
df.head()
```

```python
from datetime import datetime

lengthOfRow = 12287998

# print("hel")
# print(df[0])
# print("hel")
# print(len(df))

array_x = []
array_y = []


indices = np.arange(0, lengthOfRow, 100)
for idx in indices:
    # print(f"Value at position {idx}: {df[idx]}")

    specific_timestamp = 0 +1000 * (idx / 100)
    date_from_timestamp = datetime.fromtimestamp(specific_timestamp/1000)
    array_x.append(date_from_timestamp)
    array_y.append(df[idx].values[0])
```

```python
print(array_y[0])
print(array_y[10])
print(array_x[9])
print(array_x[10])
```

```python
print(len(array_y))
print(array_y)
```

```python
import csv

# Create and write to CSV file
with open('output-with-date.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    # Write headers
    writer.writerow(['date', 'pressure'])
    # Write data rows
    writer.writerows(zip(array_x, array_y))
```
